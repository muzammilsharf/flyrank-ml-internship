## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

This is a binary classification and scoring problem. We train a classification model to output a probability score between 0 and 1 for each web page. We use that probability score to rank our review queue from highest risk to lowest risk.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

We predict whether a web page is actively losing search traffic.

Our target label is a proxy label. It comes from the observed trend_direction column in our dataset. We create a boolean label that is marked as true when the trend direction is down. This label is based on observed historical outcomes, not an arbitrary product rule. We use this current drop in traffic as a smart substitute to predict future decline risk.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

Our primary success metric is Precision@50 (Precision at 50).

This metric evaluates the top 50 pages recommended by our model. It checks how many of those 50 pages are actually losing search traffic. Overall accuracy is a bad metric here because human reviewers only have time to check a small queue of pages. A "good" number for our project is beating the Random Forest baseline of 74 percent. This means at least 37 out of our top 50 recommendations must be genuine declining pages.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule is too rigid for search data. An if-statement that flags every page older than 180 days creates massive false alarms. It will flag older pages that are still ranking well and gaining traffic.

Search decay is messy and non-linear. A page might drop in position but still gain clicks due to seasonal demand. Machine learning evaluates content age, impressions, click rates, and ranking positions all at the same time. It weighs these signals together to separate true traffic decay from harmless daily fluctuations.
